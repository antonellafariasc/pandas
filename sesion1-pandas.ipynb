### Definiendo una nueva columna en el dataframe

```python
#clasificar a las personas del Titanic según su edad en categorías como "niño", "adolescente", "adulto" o "adulto mayor
def Age_Category(edad):
    if edad < 13:
        return "niño"
    elif 13 <= edad < 20:
        return "adolescente"
    elif 20 <= edad < 60:
        return "adulto"
    else:
        return "adulto mayor"
#aplicar la función de clasificación de edad a la columna "Age" del dataframe df_titanic2
df_titanic2["Age_Category"] = df_titanic2["Age"].apply(Age_Category)
```
